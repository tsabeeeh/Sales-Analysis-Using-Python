Performance of Orders Analysis

المنتجات الاكثر مبيعا

```python
top_products = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False)
top_products.head(10)
```

المبيعات حسب المناطق

```python
sales_by_country = df.groupby('Country')['Total price'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sales_by_country.plot(kind='barh')
plt.title("Sales by country")
plt.ylabel("country")
plt.xlabel("Total price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

تحليل المبيعات بمرور الزمن

```python
df['Order Month'] = df['Order Date'].dt.to_period('M')
sales_by_month = df.groupby('Order Month')['Total price'].sum()
plt.figure(figsize=(10,6))
sales_by_month.plot(kind='line')
plt.title("Sales Over Time")
plt.ylabel("Total price")
plt.xlabel("Month")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

تحليل فئات المنتحات

```python
sale_by_category = df.groupby('Category')['Total price'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sale_by_category.plot(kind='bar')
plt.title("Sales by category")
plt.ylabel("Total price")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

المبيعات حسب الفئة و الزمن

```python
sale_by_category_month = df.groupby([df['Order Month'],'Category'])['Total price'].sum().unstack()
sale_by_category_month.plot(kind="line",figsize=(10,6))
plt.title("Sales by category Over Time")
plt.ylabel("Total price")
plt.xlabel("Month")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

متوسط المبيعات حسب العميل

```python
sales_by_customers = df.groupby("Full Name")['Total price'].mean()
plt.figure(figsize=(10,6))
sales_by_customers.plot(kind="hist",bins=20)
plt.title("AVG Sales by Customers")
plt.ylabel("AVG Sales ")
plt.xlabel("Frequancy")
plt.grid(True)
plt.tight_layout()
plt.show()
```
